```python
import numpy as np

# 계수 행렬 A
A = np.array([[2, 4],
              [6, 2]])

# 상수 벡터 B
B = np.array([5, 7])

# 해 구하기 (X = A^{-1}B)
X = np.linalg.solve(A, B)

print("해:", X)
print(f"x = {X[0]}, y = {X[1]}")

```

\end{bmatrix}

\frac{1}{-20}
\begin{bmatrix}

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# ---- SymPy Integration ----
x = sp.Symbol('x', real=True)
f = 3*x**2 + 7*x + 6
a, b = 2, 5  # Integration bounds

F = sp.integrate(f, x)                   # Indefinite integral
area_exact = sp.integrate(f, (x, a, b))  # Exact area over [2, 5]

print("Indefinite integral F(x) =", F)
print(f"Exact area ∫ from {a} to {b} of f(x) dx =", area_exact)

# ---- Riemann Sum with Rectangles ----
N = 8   # Number of rectangles
dx = (b - a) / N
xs = np.linspace(a, b, 400)
f_np = sp.lambdify(x, f, 'numpy')

# Midpoint rule sampling
sample_x = np.linspace(a + dx/2, b - dx/2, N)
heights = f_np(sample_x)
riemann_sum = np.sum(heights * dx)

# ---- Visualization ----
plt.figure(figsize=(8, 5))
plt.plot(xs, f_np(xs), 'orange', label='f(x) = 3x² + 7x + 6')

# Draw rectangles
for xi, hi in zip(sample_x, heights):
    plt.bar(xi, hi, width=dx, alpha=0.3, edgecolor='black')

plt.axvline(a, linestyle='--', alpha=0.6)
plt.axvline(b, linestyle='--', alpha=0.6)

plt.title(f"Integration Interval [{a}, {b}] - Midpoint Riemann Sum (N={N})\n"
          f"Riemann Sum ≈ {riemann_sum:.6f},  Exact Area = {float(area_exact):.6f}")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# ---- Symbolic Integration with SymPy ----
x = sp.Symbol('x', real=True)
f = 3*x**2 + 7*x + 6
a, b = 2, 5  # Integration bounds

F = sp.integrate(f, x)                   # Indefinite integral
area_exact = sp.integrate(f, (x, a, b))  # Definite integral (exact area under the curve)

print("Indefinite integral F(x) =", F)
print(f"Exact area ∫ from {a} to {b} of f(x) dx =", area_exact)

# ---- Numerical Integration with Midpoint Riemann Sums ----
rect_counts = [2, 3, 7, 10, 100]  # Number of rectangles

xs = np.linspace(a, b, 400)
f_np = sp.lambdify(x, f, 'numpy')  # Convert symbolic function to numpy function

# Create subplots: 2 rows × 3 columns (max 6 plots)
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for idx, N in enumerate(rect_counts):
    dx = (b - a) / N
    sample_x = np.linspace(a + dx/2, b - dx/2, N)  # Midpoints
    heights = f_np(sample_x)
    riemann_sum = np.sum(heights * dx)

    ax = axes[idx]
    ax.plot(xs, f_np(xs), 'orange', label='f(x) = 3x² + 7x + 6')

    # Draw rectangles for the Riemann sum
    for xi, hi in zip(sample_x, heights):
        ax.bar(xi, hi, width=dx, alpha=0.3, edgecolor='black')

    # Vertical lines at integration bounds
    ax.axvline(a, linestyle='--', alpha=0.6)
    ax.axvline(b, linestyle='--', alpha=0.6)

    # Titles and labels
    ax.set_title(f"N = {N} Rectangles\n"
                 f"Riemann Sum ≈ {riemann_sum:.4f}, Exact Area = {float(area_exact):.4f}")
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.legend()

# Adjust layout and display plot
plt.tight_layout()
plt.show()
```
