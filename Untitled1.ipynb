$ \int \left( 3x^2 + 6x \right) \, dx $

```python
 import sympy as sp

# 변수 정의
x = sp.symbols('x')

# 적분할 함수
expr = 3*x**2 + 6*x

# 부정적분
integral = sp.integrate(expr, x)

print("∫(3x^2 + 6x) dx =", integral)
```

$
\int_{2}^{20} \left( 3x^2 + 6x \right) \, dx
$

```python
import sympy as sp

# 변수 정의
x = sp.symbols('x')

# 적분할 함수
expr = 3*x**2 + 6*x

# 정적분 (2 ~ 20)
definite_integral = sp.integrate(expr, (x, 2, 20))

print("∫ from 2 to 20 (3x^2 + 6x) dx =", definite_integral)
```

```python
# Derivative–Integral duality demo for f(x) = 3x^2 + 6x
import numpy as np
import matplotlib.pyplot as plt

# ----- define functions -----
def f(x):      # original function
    return 3*x**2 + 6*x

def F(x):      # an antiderivative (C=0)
    return x**3 + 3*x**2

# domain
x = np.linspace(0, 22, 600)

# interval for definite integral
a, b = 2, 20
xa = np.linspace(a, b, 400)

# 1) f(x) with shaded area ∫_a^b f(x) dx
plt.figure()
plt.plot(x, f(x), label=r"$f(x)=3x^2+6x$")
plt.fill_between(xa, f(xa), 0, alpha=0.2,
                 label=rf"$\int_{a}^{b} f(x)\,dx$ with $a={a},\,b={b}$")
plt.scatter([a, b], [f(a), f(b)], zorder=5)
plt.title(r"Area under $f(x)$ equals $F(b)-F(a)$  (FTC I)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# 2) F(x) showing F(b)-F(a)
Fa, Fb = F(a), F(b)
plt.figure()
plt.plot(x, F(x), label=r"$F(x)=x^3+3x^2$ (an antiderivative of $f$)")
plt.hlines([Fa, Fb], xmin=a, xmax=b, linestyles="dashed")
plt.vlines([a, b], ymin=min(Fa, Fb), ymax=max(Fa, Fb), linestyles="dotted")
plt.scatter([a, b], [Fa, Fb], zorder=5)
plt.annotate(r"$F(a)$", (a, Fa), xytext=(a+0.5, Fa+200),
             arrowprops=dict(arrowstyle="->"))
plt.annotate(r"$F(b)$", (b, Fb), xytext=(b-6, Fb+200),
             arrowprops=dict(arrowstyle="->"))
plt.title(r"$\int_{a}^{b} f(x)\,dx = F(b)-F(a)$  (Fundamental Theorem of Calculus)")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.legend()
plt.grid(True)

# 3) Verify F'(x) = f(x) (numerical derivative)
dF_num = np.gradient(F(x), x)
plt.figure()
plt.plot(x, f(x), label=r"$f(x)$")
plt.plot(x, dF_num, linestyle="--", label=r"Numerical $F'(x)$")
plt.title(r"Derivative–Integral Duality: $F'(x)=f(x)$  (FTC II)")
plt.xlabel("x")
plt.ylabel("vbalue")
plt.legend()
plt.grid(True)

plt.show()
```

$ 
\frac{d}{dx} \sin(x) = \cos(x)
 $

```python
import sympy as sp

# 변수 정의
x = sp.symbols('x')

# 함수 정의
f = sp.sin(x)

# 미분
df = sp.diff(f, x)

print("f(x) =", f)
print("f'(x) =", df)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x 범위
x = np.linspace(-2*np.pi, 2*np.pi, 600)

# 원래 함수와 도함수
y_sin = np.sin(x)
y_cos = np.cos(x)

# 그래프
plt.figure(figsize=(8, 5))
plt.plot(x, y_sin, label=r"$\sin(x)$", color="blue")
plt.plot(x, y_cos, label=r"$\cos(x)=\frac{d}{dx}\sin(x)$", color="red", linestyle="--")

# 축과 격자
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.grid(True, linestyle="--", alpha=0.6)b

# 범례와 제목
plt.legend(fontsize=12)
plt.title(r"Derivative of $\sin(x)$ is $\cos(x)$", fontsize=14)
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
# Riemann sum visualization for f(x)=3x^2+6x on [2, 20]
# Show 2, 3, 4, 5-rectangle LEFT Riemann sums, one figure per N.
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 3*x**2 + 6*x

a, b = 2.0, 20.0
xx = np.linspace(a, b, 800)

# 실제 적분값 (원시함수 F(x)=x^3+3x^2 이용)
true_area = (b**3 + 3*b**2) - (a**3 + 3*a**2)

def draw_left_riemann(N):
    dx = (b - a) / N
    lefts = a + dx*np.arange(N)       # 왼쪽 분할점들
    heights = f(lefts)
    approx = np.sum(heights * dx)     # 좌측 리만합 근사

    plt.figure()
    # 함수 곡선
    plt.plot(xx, f(xx), label=r"$f(x)=3x^2+6x$")
    # 직사각형(좌측 리만합)
    plt.bar(lefts, heights, width=dx, align='edge',
            alpha=0.3, edgecolor='k', linewidth=0.8,
            label=f"{N} rectangles (left)")
    # 장식
    plt.title(f"Left Riemann Sum with N={N}  |  Approx={approx:.2f},  True={true_area:.0f}")
    plt.xlabel("x"); plt.ylabel("y")
    plt.xlim(a-0.5, b+0.5); plt.ylim(0, f(b)*1.05)
    plt.legend(); plt.grid(True)
    plt.show()

for N in [2, 3, 4, 5, 10]:
    draw_left_riemann(N)
```
